### Importing Libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np
```

### Getting Table Content 
#### By importing BeautifulSoup and requests inorder to get the data from website

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'html.parser')
#print(soup.prettify())

#table content
table=soup.find('table',{'class':'wikitable sortable'}).tbody
rows=table.find_all('tr')
columns=[v.text.replace('\n','') for v in rows[0].find_all('th')]
#to remove the linefeed we use|n replace by '' and displaying all the rows of that particular column

```

```python
df=pd.DataFrame(columns=columns)
df
```

```python
for i in range(1,len(rows)):  
    tds=rows[i].find_all('td')
    
           
    if len(tds) ==2:
        values= [tds[0].text,'',tds[1].text,'',tds[2].text.replace('\n',''.replace('\xa0',''))]
    else:
        values=[td.text.replace('\n','').replace('\xa0','') for td in tds] 
    
    
    df=df.append(pd.Series(values,index=columns),ignore_index=True)    
    print(df)
    
```

### Reading the file


```python
pcdf=pd.read_csv('PostalCodes.txt')
pcdf.head(5)
```

### Drop the columns that have 'NaN' for both columns Borough and Neighbourhood

```python
pcdf=pcdf.replace('Not assigned', np.NaN)
pcdf=pcdf.dropna(thresh=2)
pcdf.head(10)
```

### If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
pcdf.Neighbourhood = np.where(pcdf.Neighbourhood.isnull(), pcdf.Borough, pcdf.Neighbourhood)
pcdf.Neighbourhood
pcdf.head(10)
```

### Avoiding duplicate cells by combining Neighbourhood rows to single row based on postcode

```python
pcdf = pcdf.where((pd.notnull(pcdf)), None)
pcdf=pcdf.groupby('Postcode',as_index=False).agg(lambda x:','.join(set(x.dropna())))
pcdf
```
